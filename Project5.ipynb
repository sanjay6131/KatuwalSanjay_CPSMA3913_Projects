```
from anytree import Node, RenderTree
from anytree.exporter import DotExporter
from IPython.display import display, Image
Applied_maths = Node("Applied maths Major") #root
Business_and_accounting= Node("Accounting and businees Department ", parent= Applied_maths)
Computer_Progamming = Node("Computer Progamming Department", parent=Applied_maths)
Mathematics_Department = Node("Mathematics Department", parent=Applied_maths)
Financial_Accounting = Node("Financial Accounting", parent=Business_and_accounting)
Progamming_2 = Node("Progamming 2", parent=Computer_Progamming)
Calculus= Node("Calculus",parent=Mathematics_Department)
Financial_Management = Node("Financial management", parent=Financial_Accounting)
Computer_Bus_Application = Node("Computer Bus Application",parent=Progamming_2)
Discrete_Math = Node("Discrete mathematics",parent=Calculus)
Insurance_Planning =Node("Insurance Planning And Risks",parent=Computer_Bus_Application)
DotExporter(Applied_maths).to_picture("Applied Maths Major.png")
display(Image("Applied Maths Major.png"))
```
